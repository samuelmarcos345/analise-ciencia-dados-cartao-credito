```python
categoricas = ['SEX','EDUCATION','MARRIAGE','PAY_1','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6','default payment next month']
```

Criei uma lista com as colunas categoricas.

```python
for col in categoricas:
    print("Coluna:", col)
    print(df[col].value_counts())
    print("--------------------")
```

A frequência das variáveis categóricas foi analisada utilizando o método value_counts() 
aplicado às principais colunas categóricas, como sexo, educação, estado civil e histórico de pagamentos. 
Essa análise permitiu observar a distribuição das categorias.
